```python
# Import the required modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Settings to produce plots in jupyter notebook
plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
# Read in the data
stock_data = pd.read_csv('datasets/stocks.csv', parse_dates=True, index_col=['Date']).dropna()
benchmark_data = pd.read_csv('datasets/benchmark.csv', parse_dates=True, index_col=['Date']).dropna()
```

```python
# Look at the data
print('Stocks\n')
stock_data.info()
print(stock_data.head())
print('\nBenchmarks\n')
benchmark_data.info()
print(benchmark_data.head())
```

```python
# Plotting and summarizing daily prices for Amazon and Facebook in 2016
stock_data.plot(subplots=True, title="Stock Data")
stock_data.describe()
```

```python
# Plotting and summarizing daily values for the S&P 500 from the same time period
benchmark_data.plot(title="S&P 500 2016")
benchmark_data.describe()
```

```python
# calculate daily stock_data returns
stock_returns = stock_data.pct_change()

# plot daily returns
stock_returns.plot()

# summary of daily returns
stock_returns.describe()
```

```python
# calculate daily benchmark_data returns
sp_returns = benchmark_data['S&P 500'].pct_change()

# plot the daily returns
sp_returns.plot()

# summarize the daily returns
sp_returns.describe()
```

```python
# calculate the difference in daily returns
excess_returns = stock_returns.sub(sp_returns, axis=0)

# plot the excess returns
excess_returns.plot()

# summary of the excess returns
excess_returns.describe()
```

```python
# calculate the mean of excess returns
avg_excess_return = excess_returns.mean()

# plot avg excess return
avg_excess_return.plot.bar(title="Mean of the Return Difference - FB v AMZN")
```

```python
# calculate the standard deviations
sd_excess_return = excess_returns.std()

# plot the std
sd_excess_return.plot.bar(title = "Standard Deviation of the Return Difference")
```

```python
# calculate the daily sharpe ratio
daily_sharpe_ratio = avg_excess_return.div(sd_excess_return)

# annualize the sharpe ratio
annual_factor = np.sqrt(252)
annual_sharpe_ratio = daily_sharpe_ratio.mul(annual_factor)

# plot the annualized sharpe ratio
annual_sharpe_ratio.plot.bar(title="Annualized Sharpe Ratio: Stocks v S&P 500")
```
